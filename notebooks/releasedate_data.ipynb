```python
import pandas as pd
import requests
import time

# ConfiguraciÃ³n
API_KEY = "9d73f2cc3490e088dfa75a0dbadd91d3"
BASE_URL = "https://api.themoviedb.org/3"

# Cargar dataset original
df = pd.read_csv("C:/Users/leoza/Desktop/ESPOL/ProyectoCIAP/data/clean_movies.csv")

# Columnas nuevas
df['release_date'] = ''
df['vote_average'] = None
df['vote_count'] = None

# Iterar sobre tÃ­tulos
for i, row in df.iterrows():
    title = row['title']
    print(f"Procesando: {title}")
    
    # Buscar pelÃ­cula
    response = requests.get(
        f"{BASE_URL}/search/movie",
        params={"api_key": API_KEY, "query": title}
    ).json()

    if response.get("results"):
        movie_id = response["results"][0]["id"]
        details = requests.get(
            f"{BASE_URL}/movie/{movie_id}",
            params={"api_key": API_KEY}
        ).json()
        
        df.at[i, 'release_date'] = details.get('release_date')
        df.at[i, 'vote_average'] = details.get('vote_average')
        df.at[i, 'vote_count'] = details.get('vote_count')
    else:
        print("No se encontrÃ³.")

    time.sleep(0.25)

# Guardar
df.to_csv("clean_movies_enriched.csv", index=False)
```
